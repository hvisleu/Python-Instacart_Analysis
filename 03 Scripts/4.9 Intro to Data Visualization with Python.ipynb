# Creating Line Charts

### Sampling Data

```python
# create a list holding true/false values to test the np.random.rant() <=0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# store 70% of the sample in df big
big = ords_prods_merge[dev]
```

```python
# store 30% of the sample in df small
small = ords_prods_merge[~dev]
```

```python
len(ords_prods_merge)
```

```python
len(big)+len(small)
```

```python
# create df with only columns we need
df_2 = small[['orders_day_of_week', 'prices']]
```

### Line Chart

```python
# create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# export line graph
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_dow_price.png'))
```
